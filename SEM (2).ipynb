```python
import semopy
import pandas as pd
import numpy as np
import scipy
import sympy
import sklearn
import statsmodels
import graphviz 
```

```python
from dataclasses import dataclass
data = pd.read_excel (r'/content/Excel.xlsx')
data = pd.DataFrame(data, columns= ['age','experience_in_driving','education','martial_status','children','drinking','medicine','smoke','health_problem','work_design','job_demand','repititive_work','akward_posture','knees','lower_leg','mid_back','hand_pain','shoulder_pain','neck_pain'])
data.rename(columns = {'health_problem':'tobacco_chewing'}, inplace = True)
data=data.astype(float)
data=data.fillna(0)

```

```python
Std_data=sklearn.preprocessing.scale(data)
Std_data= pd.DataFrame(Std_data, columns= ['age','experience_in_driving','education','martial_status','children','drinking','medicine','smoke','health_problem','work_design','job_demand','repititive_work','akward_posture','knees','lower_leg','mid_back','hand_pain','shoulder_pain','neck_pain' ])

#d=sklearn.preprocessing.StandardScaler.fit_transform(df['age'])
```

```python
# Specify the model relations using the same syntax given before
model_spec = """
  # measurement model
    personal_factors =~  age +  experience_in_driving+martial_status+education+children
    habits =~ medicine + tobacco_chewing+drinking
    work =~ work_design + job_demand + repititive_work + akward_posture
    WRMSDs =~ knees + lower_leg + mid_back+shoulder_pain+neck_pain
  #correlation
    #age~~experience_in_driving
    #drinking~~medicine
    #drinking~~health_problem
    #medicine~~health_problem
    personal_factors~~habits
    personal_factors~~work
    #personal_factors~~WRMSDs
    habits~~work
    #habits~~WRMSDs
    #work~~WRMSDs

  # regressions
    WRMSDs ~ personal_factors+habits + work
"""
# Instantiate the model
model = semopy.Model(model_spec)
# Fit the model using the data
print(model.fit(data))
print(semopy.calc_stats(model).T)
# Show the results using the inspect method
model.inspect(mode='list', std_est=True)
```

```python
import graphviz
from google.colab import files
semopy.semplot(model, 'SEM.png', std_ests=True, plot_covs=True)
files. download("SEM.png")
```
